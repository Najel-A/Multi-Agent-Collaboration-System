## 04 — Evaluation

This notebook is for producing report-ready evaluation outputs.

- Classification: metrics + confusion matrix (from saved test metrics)
- Generation: qualitative examples (prompt → model output vs target)

It’s intentionally modular so you can add new metrics later.


```python
from pathlib import Path
import json

THIS_NOTEBOOK_DIR = Path.cwd()
ARTIFACTS = (THIS_NOTEBOOK_DIR.parent / "artifacts").resolve()

metrics_path = ARTIFACTS / "classifier_distilbert" / "metrics_test.json"
if metrics_path.exists():
    metrics = json.loads(metrics_path.read_text())
    print(json.dumps({k: metrics[k] for k in ["accuracy", "precision_macro", "recall_macro", "f1_macro"] if k in metrics}, indent=2))
else:
    print("No classifier metrics found at", metrics_path)
```
